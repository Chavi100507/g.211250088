```python
def cariBiner(dataTerurut, nilaiDicari):
    batasKiri = 0
    batasKanan = len(dataTerurut) - 1

    while batasKiri <= batasKanan:
        indeksTengah = (batasKiri + batasKanan) // 2
        nilaiTengah = dataTerurut[indeksTengah]

        if nilaiTengah == nilaiDicari:
            return indeksTengah

        if nilaiTengah > nilaiDicari:
            batasKanan = indeksTengah - 1
        else:
            batasKiri = indeksTengah + 1

    return -1

daftarAngka = [1, 3, 5, 7, 9, 11, 13, 15]
targetAngka = 7

hasil = cariBiner(daftarAngka, targetAngka)

if hasil != -1:
    print(f"input: {daftarAngka}, target: {targetAngka}")
    print(f"output: elemen ditemukan di index {hasil}")
else:
    print("output: elemen tidak ditemukan")
```

```python
angkaInput = int(input("Masukkan angka faktorialnya: "))

def hitungFaktorial(nilai):
    if nilai <= 1:
        return 1
    else:
        return nilai * hitungFaktorial(nilai - 1)

print(hitungFaktorial(angkaInput))
```
